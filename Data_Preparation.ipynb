## Feature selection 

1. Select Percentile
2. Generic Univariate Selet
3. Select KBest
4. Select From Model

```python
from sklearn.feature_selection import SelectPercentile
from sklearn.feature_selection import chi2,f_classif
from sklearn.datasets import load_breast_cancer

data=load_breast_cancer()
X=data.data
Y=data.target
print(X.shape)
Sp=SelectPercentile(score_func=chi2,percentile=20)

X=Sp.fit_transform(X,Y)

print(X.shape)

# print(Y)
print(Sp.get_support()) # get the feature that is included


```

```python
from sklearn.feature_selection import GenericUnivariateSelect

data=load_breast_cancer()
X=data.data
Y=data.target
print(X.shape)

Gen=GenericUnivariateSelect(score_func=chi2,param=4,mode='k_best')
X=Gen.fit_transform(X,Y)

print(X.shape)
print(Gen.get_support())
```

```python
from sklearn.feature_selection import SelectKBest

data=load_breast_cancer()
X=data.data
Y=data.target
print(X.shape)

Gen=SelectKBest(score_func=chi2,k=4)
X=Gen.fit_transform(X,Y)

print(X.shape)
print(Gen.get_support())
```

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.linear_model import LinearRegression


model=LinearRegression()

data=load_breast_cancer()
X=data.data
Y=data.target
print(X.shape)

Gen=SelectFromModel(estimator=model,max_features=5)
X=Gen.fit_transform(X,Y)

print(X.shape)
print(Gen.get_support())
```

```python
from sklearn.preprocessing import Normalizer
data=load_breast_cancer()
X=data.data
print(X)
Y=data.target
scalre=Normalizer(norm='max')# devide each row in the max in that row
X=scalre.fit_transform(X)
print(X)
```

```python
from sklearn.preprocessing import MaxAbsScaler
data=load_breast_cancer()
X=data.data
print(X)
Y=data.target
scalre=MaxAbsScaler() ## devide each col in the max in that column
X=scalre.fit_transform(X)
print(X)
```

```python
from sklearn.preprocessing import FunctionTransformer
data=load_breast_cancer()
X=data.data
print(X)
Y=data.target
scalre=FunctionTransformer(func=lambda x:x/2,validate=True)
X=scalre.fit_transform(X)
print(X)
```

```python
from sklearn.preprocessing import Binarizer
data=load_breast_cancer()
X=data.data
print(X)
Y=data.target
scalre=Binarizer(threshold=10000000) # if the value is greater than thereshold => 2 else => 0
X=scalre.fit_transform(X)
print(X)
```

## Data Encoding 

```python
from sklearn.preprocessing import LabelEncoder
dataset =pd.read_csv('House_Rent_Dataset.csv')
# print(dataset)

lbl=LabelEncoder()


dataset['Posted On']=lbl.fit_transform(dataset['Posted On'])


dataset['Floor']=lbl.fit_transform(dataset['Floor'])


dataset['Area Type']=lbl.fit_transform(dataset['Area Type'])


dataset['Area Locality']=lbl.fit_transform(dataset['Area Locality'])

dataset['City']=lbl.fit_transform(dataset['City'])

dataset['Furnishing Status']=lbl.fit_transform(dataset['Furnishing Status'])


dataset['Tenant Preferred']=lbl.fit_transform(dataset['Tenant Preferred'])


dataset['Point of Contact']=lbl.fit_transform(dataset['Point of Contact'])
print(type(dataset))
# print(dataset)
```

```python

simpleImputer=SimpleImputer(missing_values=np.nan,strategy='mean')
dataset[dataset.columns]=simpleImputer.fit_transform(dataset)

simpleImputer=SimpleImputer(missing_values=0,strategy='mean')
dataset[dataset.columns]=simpleImputer.fit_transform(dataset)

print(type(dataset))

X=np.array(dataset.drop(columns='Rent'))
print(X.shape)
Y=np.array(dataset['Rent'])

X=StandardScaler().fit_transform(X)


X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=.3,shuffle=True)


print(X_train.shape)
print(Y_train.shape)

print(X_test.shape)
print(Y_test.shape)
```
